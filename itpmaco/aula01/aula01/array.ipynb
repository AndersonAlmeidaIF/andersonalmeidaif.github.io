- Gerando uma matriz de ordem n:

```python
n = 5

C = [ [],
      [],
      [],
      [],
      [] ]
```

- Gerando uma matriz nula:

```python
C = [ [],[],[],[],[] ]

for i in range(n):
    for j in range(n):
        C[i].append(0)
```

```python
for i in range(n):
    for j in range(n):
        print(C[i][j], " ")
    print("\n")
```

- Preenchendo com valores aleatórios:

```python
C = [ [],[],[],[],[] ]

for i in range(n):
    for j in range(n):
        C[i].append(random.randint(0,9))
```

```python
for i in range(n):
    for j in range(n):
        print(C[i][j], " ")
    print("\n")
```

- Preenchendo e exibindo simultaneamente:

```python
C = [ [],[],[],[],[] ]

for i in range(n):
    for j in range(n):
        C[i].append(0)
        print(C[i][j], " ")
        
    print("\n")
```

- Gerando uma matriz identidade:

```python
C = [ [],[],[],[],[] ]

for i in range(n):
    for j in range(n):

        if i == j:
            C[i].append(1)
        else:
            C[i].append(0)
            
        print(C[i][j], " ")
        
    print("\n")
```
